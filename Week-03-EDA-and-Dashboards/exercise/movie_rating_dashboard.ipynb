```python
import pandas as pd
import altair as alt
import plotly.express as px

df= pd.read_csv('data/movie_ratings.csv')

#print(df.info())
print(df.tail())
```

```python
df['genres'].value_counts()
```

```python
genre_counts=df['genres'].value_counts().reset_index()
genre_counts.columns=['genre','count']
genre_counts

```

```python
fig = px.bar(
    genre_counts,
    x='count',
    y='genre',
    orientation='h',
    title="Movie Genres by Ratings",
    labels={'count': 'Number of Ratings', 'genre': 'Genre'}
)
fig.show()
```

**Most rated Genres:**

-Drama leads with close to 40k ratings and Comedy is also up there with around 30k ratings

**Moderately rated Genres:**

-Action and Thriller with ratings from 21-25k
-Romance, Adventure, and Sci-Fi with ratings from 12k-20

**Least Rated Genres:**

-Documenttary being the lowest reviewed with less than 1k reviews
-Fantasy & Film Noir with a little higher than 1k


```python
genre_rating=df.groupby('genres')['rating'].mean().reset_index()
genre_rating=genre_rating.sort_values(by='rating', ascending=False)
genre_rating
```

```python
fig2= px.bar(
    genre_rating,
    x='rating',
    y='genres',
    orientation='h',
    title= 'Average Rating Per Genre',
    labels={'rating': 'Average Rating', 'genres':'Genre'}
)
fig2.update_layout(
    xaxis=dict(range=[3.1,4.0])
)
fig2.show()
```

**Highest rated Genres:**

-Film Noir(3.92),War(3.81) have the highest

-while they arent the most popular their fans love its movies very highly

**Lowest Rated Genres:**

-Fantasy (3.21), Horror (3.29), and Children (3.35) have the lowest ratings

```python
df['year'].value_counts()
```

```python
year_rating=df.groupby('year')['rating'].mean().reset_index()
year_rating
```

```python
fig3 = px.line(
    year_rating,
    x='year',
    y='rating',
)
fig3.update_xaxes(dtick=10)
fig3.show()
```

- Ratings were higher in older movies (1930–1960), with peaks like 1942 (4.4)
- From the 1970s forwards, ratings show a downward trend, reaching around 3.3 by the 1990
- Overall, older films seem to be rated more favorably than newer one

```python
#df['title'].value_counts()
movie_stats=df.groupby('title').agg(
    avg_rating=('rating', 'mean'),
    num_ratings=('rating', 'count')
).reset_index()
movie_stats
```

```python
best_50=(movie_stats[movie_stats['num_ratings']>= 50].sort_values(by='avg_rating', ascending=False).head(5))
best_50
```

```python
best_150=(movie_stats[movie_stats['num_ratings']>= 150].sort_values(by='avg_rating', ascending=False).head(5))
best_150
```

```python
fig4= px.bar(
    best_50,
    x='avg_rating',
    y='title',
    orientation='h',
    title= 'Top 5 Movies With At Least 50 Reviews ',
    text='avg_rating',
    labels={'avg_rating': 'Average Rating'}
)

fig4.update_xaxes(range=[4.3, 4.55])
fig4.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig4.show()
```

The best-rated movies all have very high averages (4.45–4.49).

A Close Shave”(1995) leads with the highest rating (4.49).

```python
fig4= px.bar(
    best_150,
    x='avg_rating',
    y='title',
    orientation='h',
    title= 'Top 5 Movies With At Least 150 Reviews ',
    text='avg_rating',
    labels={'avg_rating': 'Average Rating'}
)

fig4.update_xaxes(range=[4.3, 4.55])
fig4.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig4.show()
```

- When we raise the bar to 150 reviews Shawshank Redemption takes a spot in the top 5

- This pushes out Wallace & Gromit, which was in the top 5 with the 50 review limit

- The other four movies stay the same, showing they are strong no matter the cutoff
